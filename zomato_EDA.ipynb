```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('zomato.csv',encoding='latin-1')
df.head()
```

```python
df_countrycode=pd.read_excel('Country-Code.xlsx')
```

```python
df_countrycode.head()
```

```python
df_countrycode.columns
```

```python
final_df=pd.merge(df,df_countrycode,how='left',on='Country Code')
```

```python
final_df.head(2)
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
country_name=final_df.Country.value_counts().index
```

```python
country_name
```

```python
country_value=final_df.Country.value_counts().values
```

```python
country_value
```

```python
# pie chart-top three country that uses zomato
plt.pie(country_value[:3],labels=country_name[:3],autopct='%1.2f%%')
```

# observation:
zomato maximum transaction From india,after that USA , then From UK.

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating_count'})
```

```python
ratings
```

# observation

1.when rating is between 4.5 to 4.9 -->Excellent
2.when rating is between 4 to 4.4 -->Very good
3.when rating is between 3.5 to 3.9 -->good
4.when rating is between 2.5 to 3.4 -->Average
5.when rating is between 1.8 to 2.4 -->poor


```python
ratings.head()
```

```python
import matplotlib

matplotlib.rcParams['figure.figsize']=(12,6)

sns.barplot(x='Aggregate rating',y='Rating_count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating_count',data=ratings,hue='Rating color',palette=['blue','red','orange','yellow','green','green'])
```

#observation:
    1.Not Rated count is very high
    2.Maximum number of rating are between 2.5 to 3.4

```python
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
